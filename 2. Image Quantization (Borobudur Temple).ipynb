# 1. Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

# 2. Import Picture

```python
img = plt.imread("data/Borobudur_Temple.jpeg")
img.shape
```

```python
plt.imshow(img)
```

# 3. Converting Picture to Data 

```python
img = img.reshape(-1, 3)
img
```

```python
len(img)
```

```python
len(np.unique(img, axis=0))
```

# 4. Cluster 34417 colours into 10 colour palette

```python
kmeans = KMeans(n_clusters=10, random_state=42, n_jobs=-1)
labels = kmeans.fit_predict(img)
```

```python
labels
```

```python
palette = kmeans.cluster_centers_.astype(int)
palette
```

```python
quantized_img = palette[labels].reshape(364, 650, 3)
quantized_img
```

```python
plt.imshow(quantized_img)
```
